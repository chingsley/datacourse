```python
df_smurfs_raw = pd.read_html('https://en.wikipedia.org/wiki/List_of_The_Smurfs_characters', header=0)[4]

df_smurfs_raw
```

```python
df_smurfs_raw = df_smurfs_raw.set_index('Name').drop('(should not be counted). Don Smurfo', axis=0)
df_smurfs_raw
```

```python
df_smurfs_raw.reset_index(inplace=True)
```

```python
df_smurfs_raw.columns
```

```python
df_smurfs_description = df_smurfs_raw.drop_duplicates(subset=['Name'], keep='last')['Original French name']
df_smurfs_description
```

```python
df_smurfs = df_smurfs_raw.drop_duplicates(subset='Name', keep='first')

df_smurfs
```

```python
list(df_smurfs_description)
```

```python
df_smurfs['description'] = list(df_smurfs_description)

df_smurfs
```

```python
df_smurfs.columns = ['alias', 'name', 'movies', 'voicedBy', 'description']

df_smurfs.head()
```

```python
def format_vb(vb):
    res = vb.split(')')
    res2 = [item+')' if '(' in item else item for item in res if item != '']
    res3 = ['unknown' if item == 'N\\A' else item for item in res2]
    return res3
```

```python
df_smurfs.name.fillna('unknown', inplace=True)
df_smurfs.voicedBy.fillna('unknown', inplace=True)
df_smurfs['movies'] = df_smurfs['movies'].str.split('+')
df_smurfs['voicedBy'] = df_smurfs['voicedBy'].apply(format_vb)
df_smurfs['alias'] = df_smurfs['alias'].apply(lambda x: x.split('. ')[1])
```
